```python
from time import sleep
from bs4 import BeautifulSoup
from splinter import Browser
```

```python
executable_path = {'executable_path':'/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path)
f = open('data.csv', 'w')
f.write('CRNumber,Description,Victim,Reported,Location\n')

# Get first page
url = 'http://www.columbuspolice.org/Reports/Results?from=11/1/2018&to=11/1/2019&loc=all&types=9'
browser.visit(url)
html = BeautifulSoup(browser.html, 'html.parser')
rows = html.find('table').findChildren('tr')
for row in rows[1:-2]:
    data = [x.get_text() for x in row.findChildren('td')]
    f.write(data[0] + ',' + data[1] + ',' + data[2] + ',' + data[3] + ',' + data[4] + '\n')
    
# Get subsequent pages
# last run crashed at page 107
for i in range(107, 500):
    print('\r%d/%d' % (i, 500), end='')
    links = browser.find_by_tag('a')
    first = True
    for link in links:
        if i == 10 and link.value == '...':
            # look for the first '...'
            link.click()
        if i % 10 == 0 and link.value == '...':
            # look for the second '...'
            if first:
                first = False
            else:
                link.click()
        elif link.value == str(i):
            link.click()
    sleep(3) # wait for table to reload

    html = BeautifulSoup(browser.html, 'html.parser')
    rows = html.find('table').findChildren('tr')
    for row in rows[1:-2]:
        data = [x.get_text() for x in row.findChildren('td')]
        f.write(data[0] + ',' + data[1] + ',' + data[2] + ',' + data[3] + ',' + data[4] + '\n')
    
f.close()
browser.quit()
```
